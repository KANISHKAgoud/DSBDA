```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import PowerTransformer

# Load the dataset
file_path = r"C:\Users\comp\Documents\DSBDA PRACTICALS\DSBDA\2nd_studentPerformance\StudentsPerformance.csv"
df = pd.read_csv(file_path)

print("=== First 5 Rows ===")
print(df.head())

print("\n=== Dataset Info ===")
print(df.info())

# 1. MISSING VALUES & INCONSISTENCIES
print("\n=== Missing Values ===")
print(df.isnull().sum())

# Handling missing values: Fill numeric with mean, categorical with mode
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("\n=== Missing Values After Imputation ===")
print(df.isnull().sum())

# 2. OUTLIER DETECTION & HANDLING (Using Z-score method)
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("\n=== Outlier Detection using Z-score ===")
z_scores = np.abs(stats.zscore(df[numeric_cols]))
outliers = (z_scores > 3).sum(axis=0)
print("Number of outliers per column:\n", dict(zip(numeric_cols, outliers)))

# Removing rows with outliers (if any z-score > 3)
df_no_outliers = df[(z_scores < 3).all(axis=1)]

print(f"\n=== Rows Before Removing Outliers: {df.shape[0]}")
print(f"=== Rows After Removing Outliers: {df_no_outliers.shape[0]}")

# 3. DATA TRANSFORMATION
# Let's choose a skewed column and apply a log transformation

# Checking skewness
print("\n=== Skewness Before Transformation ===")
skewed_cols = df_no_outliers[numeric_cols].apply(lambda x: x.skew())
print(skewed_cols)

# Selecting one skewed column for log transformation
skewed_col = skewed_cols[skewed_cols > 1].index[0] if (skewed_cols > 1).any() else numeric_cols[0]
print(f"\nApplying log transformation on: {skewed_col}")

# Add 1 to avoid log(0)
df_no_outliers[f"{skewed_col}_log"] = np.log1p(df_no_outliers[skewed_col])

# Compare original and transformed
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(df_no_outliers[skewed_col], kde=True)
plt.title(f"Original Distribution of {skewed_col}")

plt.subplot(1, 2, 2)
sns.histplot(df_no_outliers[f"{skewed_col}_log"], kde=True, color='green')
plt.title(f"Log-Transformed Distribution of {skewed_col}")
plt.tight_layout()
plt.show()

# Final Data Summary
print("\n=== Final Cleaned Data Overview ===")
print(df_no_outliers.describe())
```
