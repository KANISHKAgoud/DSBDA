```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
file_path = r"C:\Users\comp\Documents\DSBDA PRACTICALS\DSBDA\9th_practical\TitanicDataset.csv"
df = pd.read_csv(file_path)

# Clean column names (strip leading/trailing spaces)
df.columns = df.columns.str.strip().str.lower()

# Display dataset info
print("Dataset Info:")
print(df.info())

# Check first few rows
print("\nFirst 5 rows:")
print(df[['sex', 'age', 'survived']].head())

# Drop rows with missing age or sex
df_clean = df.dropna(subset=['age', 'sex', 'survived'])

# Convert 'survived' to categorical for better plot labels
df_clean['survived'] = df_clean['survived'].map({0: 'Not Survived', 1: 'Survived'})

# Plot boxplot: Age distribution by sex and survival
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=df_clean, palette='Set2')
plt.title('Box Plot of Age by Gender and Survival Status')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survival Status')
plt.show()

# -----------------------------
# ðŸ“Œ Observations:
# -----------------------------
# 1. Females who survived tend to be younger on average than those who did not.
# 2. Males who survived had a slightly wider range of ages than females.
# 3. Median age of non-surviving males is higher compared to surviving males.
# 4. Many children (age < 10) are observed among the survivors, particularly females.
```
